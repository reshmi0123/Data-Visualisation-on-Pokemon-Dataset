# Data Visualization of Pokémon Data with Python and Seaborn

## Importing the libraries and data (Pandas, Seaborn and Matplotlib)



```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
import seaborn as sns

df = pd.read_csv(r"D:\ACADEMICS\DA\Pokémon.csv", encoding='latin-1', index_col=0)

df.head()
```

## Plotting of Attacks and Defenses

```python
sns.lmplot(x = 'Attack', y = 'Defense', data = df)
```

#### Correlation between Attack & Defense of each Pokémon.
#### Inference :-
##### Almost more than 60% Pokémons have their attack within the range of 60-80.
##### The average defense range is 50-85.

## Scatter Plotting of Attack and Defenses ( Without the Regresssion Line )

```python
sns.lmplot(x='Attack', y='Defense', data=df,fit_reg=False, hue='Stage')
```

### Correlation between Attack & Defense of each Pokémon.
#### Inference :-
##### We see that the Stage 1 Pokémons have mostly lesser attack (40-60 pts) but has higher defense (50-100 pts)
##### Stage 2 Pokémons have moderately higher attack (80-120 pts) with higher defense (70-125 pts)
##### Stage 3 Pokémons have scattered defense & attack pts varing from individuals

## Box Plots of Pokémon Data

#### For viewing the combat stats of Pokémon

```python
sns.boxplot(data=df)
```

#### We can remove the Stage & Legendary attributes since these do not contribute to combat stats.
#### We can remove the Total since we have individual stats.

#### Individual combat stats of every type of Pokémon

```python
# Pre-format DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis = 1)
 
# New boxplot using stats_df
sns.boxplot(data = stats_df)
```

### Inference :-
##### We see that almost every type of Pokémon have an average combat pts within the range of 50-100 pts.
##### Better to choose a Speed Atk type or Speed type Pokémon for Duel for slightly higher advantage.

## Violin plots 

#### Distribution of Attack by Pokémon's primary type

```python
# Set theme
sns.set_style('whitegrid')
 
# Violin plot
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors) # Set color palette
plt.rcParams['figure.figsize'] = 25, 25
```

#### Pokémon fans might find something quite jarring about that plot: The colors are nonsensical initially.
#### So we have set each type with there refering custom colour.

## Swarm plot

####  Swarm plot with Pokémon color palette

```python
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
plt.rcParams['figure.figsize'] = 5, 5
```

#### This visualization will show each 151 Pokémon in our dataset based on its type.

## Pie Chart

#### Visualisation of Pokémon based on 'Type 1' i.e. basic nature.

```python
df['Type 1'].value_counts().plot.pie(shadow=False,figsize=(10,8))

plt.show()
```

#### Inference :-
##### We see that 25% of all Pokémon is basically Water type & Normal type, consequently decreasing to Posion type, Fire type ultimately to Ice, Fairy types which are rare.

#### Visualisation of Pokémon based on 'Type 2' i.e. secondary nature.

```python
df['Type 2'].value_counts().plot.pie(shadow=False,figsize=(10,8))

plt.show()
```

### Inference :-
##### We see that 35% of all Pokémon has secondary type as Flying type & Posion type, consequently decreasing to Ground type, Psychic type ultimately to Fighting, Steel types which are rare.

## Creating a Plot with Violin & Swarm Plots

### Correlation between Attack by type & Pokémon type


```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1',y='Attack', data=df, inner=None, palette=pkmn_type_colors)

sns.swarmplot(x='Type 1', y='Attack', data=df, color='k', alpha=0.7) 

plt.title('Attack by Type')
```

### Inference :-
#### We have a chart that tells us how Attack values are distributed across different Pokémon types.

### Combining few attributes for a New DataSet 

```python
melted_df = pd.melt(stats_df, id_vars = ["Name", "Type 1", "Type 2"], var_name = "Stat") 
melted_df.head()
```

## Creating a Swarm plot based on each Pokémon type and their diffrent attributes

```python
plt.figure(figsize = (10,10))

sns.swarmplot(x = 'Stat', y = 'value', data = melted_df, hue = 'Type 1', dodge = True, palette = pkmn_type_colors) 
 

plt.ylim(0, 260)
 
plt.legend(bbox_to_anchor = (1, 1), loc = 2)
```

#### Inference :-
##### We have a chart that tells us how different Pokémons perform based on their types.

## Creating a Heat Map of the Pokémon Data

```python
plt.figure(figsize = (10,10))
# Calculating any correlations

corr = stats_df.corr()
 
# Creating the Heat Map
sns.heatmap(corr)
```

#### Heatmaps help us visualize matrix-like data.

## Bar Plots

```python
plt.figure(figsize = (10,10))
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

#### Inference :-
##### This shows that more than 25% of Pokémons are of Water type, more than 20% of Pokémons are of Normal types.
##### The least type being the Fairy type <5% .

## Factor Plot

### Factor plots make it easy to separate plots by categorical classes.


```python
plt.figure(figsize = (10,10))

g = sns.factorplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  
                   col='Stage',  
                   kind='swarm') # Swarmplot
 

g.set_xticklabels(rotation=-45)
```

#### Inference :-
##### This shows the distribution of different types of Pokémons based on their Stages Vs their Attack ability.

## Creating a Histogram of Pokémon Data.

#### Histograms allow us to plot the distributions of numeric variables.

```python
plt.figure(figsize = (10,10))
sns.distplot(df.Attack)
```

#### Inference :-
##### This shows the distribution of  Pokémons based on their Attack ability.

## Creating a Density Plot of Pokémon Data

#### Density plots display the distribution between two variables.

```python
plt.figure(figsize = (10,10))

sns.kdeplot(df.Attack, df.Defense)
```

#### Inference :-
##### This shows the distribution of Pokémons based on their Attack ability Vs Defense ability.
